```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]

# X.head()
# Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# The random_state parameter in the train_test_split function controls the shuffling applied to the data before the split. By setting a specific value for random_state, you ensure that the split is reproducible every time you run the code.

# How It Works
# When you set random_state to an integer (e.g., random_state=42), it acts as a seed for the random number generator. This seed ensures that the same "random" process is repeated consistently across different runs, leading to the same data being selected for training and testing each time.
# If random_state is not set (or set to None), the split will be different every time you run the code, since the random number generator will have a different seed each time.

# Why Use random_state?
# Reproducibility: If you're sharing your code or collaborating with others, setting random_state ensures that everyone gets the same results.
# Consistency in Testing: When tuning models, having a consistent train-test split allows you to fairly compare the performance of different models.
# The specific value you choose for random_state (like 42) doesnt matterits just a seed that ensures consistency.
```

```python
# Standardization of the dataset.
# Reason behind standardization of the dataset
# Standardization of data in preprocessing is a technique used to scale the features of a dataset so that they have a mean of 0 and a standard deviation of 1. This is particularly important when features have different units or scales, which can impact the performance of machine learning algorithms that rely on the distance between data points, such as k-nearest neighbors, support vector machines, and gradient descent-based methods.

# Benefits of Standardization
# Improved Model Performance: Many algorithms perform better when features are on a similar scale.
# Faster Convergence: In optimization algorithms like gradient descent, standardization can lead to faster convergence.
# Interpretability: Standardization allows you to compare the impact of different features on the model more easily, since they are all on the same scale.

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# In scikit-learn, the StandardScaler is commonly used for standardizing features by removing the mean and scaling them to unit variance. The methods fit_transform() and transform() play key roles in this process but serve different purposes. Heres how they differ:

# 1. fit_transform() Method:
# Purpose: This method combines both fitting the scaler to the data and then transforming the data based on that fit. Its typically used on the training dataset.
# Steps Involved:
# Fit: It calculates the mean and standard deviation of the features in the dataset (i.e., finds the parameters needed for standardization).
# Transform: It standardizes the dataset using the calculated mean and standard deviation.
# Use Case: Use this method when you are working with the training data, as it needs to learn the scaling parameters (mean and standard deviation) from this data.

# 2. transform() Method:
# Purpose: This method is used only to apply the scaling transformation using the parameters (mean and standard deviation) already learned by the fit() method. It does not re-calculate these parameters; it simply uses the ones determined during the fitting process.
# Steps Involved:
# It takes the pre-learned mean and standard deviation (from the training data) and applies them to scale new data (e.g., the test set).
# Use Case: Use this method when you want to transform your test data or any new data using the same parameters that were determined from the training data.

# Why the Difference Matters
# The separation between fit_transform() and transform() is crucial because:

# Avoiding Data Leakage: When you fit the scaler on the test data (or new data), it can cause data leakage by letting the model see information it should not have during training, leading to overly optimistic performance.
# Consistency: The mean and standard deviation should be consistent across the train and test datasets for proper scaling. Therefore, you only fit on the training data and transform both the training and test data with the same scaler.
# Summary
# fit_transform(): Fits the scaler (learns mean and standard deviation) and then transforms the data. Used on training data.
# transform(): Uses the parameters learned by fit() to scale the data. Used on test data or any new data.

# X_train
X_test
```

Model Training 

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, Y_train)
print(regression.coef_)
# regression.coef_:
# The coef_ attribute contains the coefficients (also called weights) of the linear regression model. These coefficients represent the relationship between each feature in your input data and the target variable.
# The coef_ attribute contains the values [b_1, b_2, ..., b_n]. Each coefficient indicates the impact of a corresponding feature on the target variable.
print(regression.intercept_)
# regression.intercept_:
# The intercept_ attribute contains the intercept (also called the bias term) of the linear regression model.
# In the linear regression equation:
#  . This is the point where the regression line crosses the y-axis when all feature values are zero.

print(regression.get_params())
# regression.get_params():
# The get_params() method is a built-in function in scikit-learn that returns the parameters (also called hyperparameters) of the model as a dictionary.
# These parameters control the behavior of the model and can be set when initializing the model.
# For instance, in a LinearRegression model, some common hyperparameters include fit_intercept, normalize, and n_jobs.
```

```python
reg_pred = regression.predict(X_test)

plt.scatter(x=Y_test, y=reg_pred)
plt.xlabel('Y test data')
plt.ylabel('Predicted values by the linear regression ')
plt.show()
```

```python
residuals = Y_test - reg_pred
# In machine learning, residuals are the differences between the observed values and the values predicted by a model. They are a key component in assessing the performance of regression models.

# Understanding Residuals
# Definition:

# Residual = Observed Value - Predicted Value
# For each data point, the residual measures how far off the models prediction is from the actual value.
# Mathematical Representation:
 
# Purpose:

# Residuals are used to evaluate the fit of a regression model. Small residuals indicate that the model's predictions are close to the actual values, while large residuals suggest that the model is not performing well.
# Residual Analysis:

# Residual Plot: A scatter plot of residuals versus predicted values or versus independent variables can help diagnose issues like non-linearity, heteroscedasticity (non-constant variance of residuals), and outliers.
# Mean Residual Error: In a well-fitting model, the mean of the residuals should be close to zero.
# Applications:

# Model Diagnostics: Checking residuals helps to understand if the model meets the assumptions of linear regression, such as linearity and homoscedasticity.
# Improving Models: If residuals show patterns or trends, it might indicate that the model could be improved or that additional features are needed.
residuals
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print(mean_squared_error(Y_test, reg_pred))
print(mean_absolute_error(Y_test, reg_pred))
print(np.sqrt(mean_squared_error(Y_test, reg_pred)))
```

R square and adjusted R square

```python
# R-Squared (R) and Adjusted R-Squared are statistical metrics used to evaluate the performance of a regression model, especially in linear regression. Both metrics assess how well the independent variables explain the variability in the dependent variable, but they have key differences.

# 1. R-Squared (R)
# Definition: R-squared, also known as the coefficient of determination, measures the proportion of the variance in the dependent variable that is predictable from the independent variables.

# Formula: R^2 = 1 - (Sum of squared Residuals SSR/Total Sum of Squares SST)

# Interpretation:
# R = 1: The model perfectly explains the variability in the dependent variable.
# R = 0: The model does not explain any variability in the dependent variable.
# Higher R values indicate a better fit.
# Limitation: R-squared tends to increase as you add more independent variables to the model, even if those variables do not improve the models accuracy. This can lead to overfitting.

# 2. Adjusted R-Squared
# Definition: Adjusted R-squared modifies the R-squared value by accounting for the number of independent variables in the model. It adjusts for the addition of variables that do not improve the model, thereby giving a more accurate measure of model performance.

# Formula: Adjusted R^2 = 1 - (((1 - R^2)*(n - 1))/(n - p - 1))
# Where:
# n = number of data points (sample size)
# p = number of predictors (independent variables)

# Interpretation:
# Adjusted R can be lower than R because it penalizes the model for adding irrelevant predictors.
# It provides a more reliable metric when comparing models with different numbers of independent variables.
# Advantage: Adjusted R-squared only increases if the new predictor improves the model more than would be expected by chance. It decreases if the predictor does not contribute significantly.

# Example
# Suppose you have a regression model with an R of 0.90. This suggests that 90% of the variance in the dependent variable is explained by the model.
# However, if the adjusted R is 0.85, it indicates that the models performance is slightly overestimated when relying solely on R, possibly due to some irrelevant predictors.
# Conclusion
# R-squared gives a basic measure of model fit, while Adjusted R-squared provides a more realistic evaluation, particularly when dealing with multiple predictors.
# In practice, adjusted R-squared is often preferred when building models with multiple variables, as it gives a better picture of the models true predictive power.

from sklearn.metrics import r2_score
score = r2_score(y_true=Y_test, y_pred=reg_pred)
print(score)
```

```python
adjusted_r2 = 1 - (1 - score)*((len(Y_test) - 1)/(len(Y_test) - X_test.shape[1] - 1))
print(adjusted_r2)
```
